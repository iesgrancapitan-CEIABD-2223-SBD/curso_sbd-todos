```python
# Code to read csv file into Colaboratory:

# instalamos la libreria de acceso a Drive

!pip install -U -q PyDrive

# importamos pandas para manipular csv

import pandas as pd

# importamos la libreria para manipular cadenas y usar 
# expresiones regulares

import re 

# importamos librerías de acceso a Drive

from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials

# Authenticate and create the PyDrive client.

auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)
```

copia el enlace al fichero csv en la variable link. Extraemos el id del fichero usando la librería de Python para expresiones regulares.

```python
link = "https://drive.google.com/file/d/1EdfQ42me8sP3G3JdPMd8uTIMYH5Qapv9/view?usp=sharing"

# usamos la libreria re de Python para buscar y extraer subcadenas. En big data es una herramienta fundamental.

# paso 1, definimos el regex a usar

p = re.compile(r".*\/d\/(\w+)\/view.+")

# paso 2, aplicamos el regex sobre la cadena de la variable link

resultado =  p.search(link)

#paso 3, con la subcadena extraida de la variable link definimos variable id 
#y la mostramos por pantalla para comprobar que se ha extraído correctamente.

id = resultado.group(1)
print(resultado.group(1))

```

una vez tenemos el id, usamos la libreria de python para llamar a ese fichero y se lo pasamos a read_csv que lo carga en la variable df3.

```python
downloaded = drive.CreateFile({'id':id}) 
downloaded.GetContentFile('ejemplo.csv')  
df3 = pd.read_csv('ejemplo.csv')
# Los datos están ahora almacenados en un Pandas Dataframe
# el método head muestra los 4 ó 5 primeros elementos del csv, lo dejamos comentado
#df3.head()
# muestra todo el contenido del fichero.
df3
```
